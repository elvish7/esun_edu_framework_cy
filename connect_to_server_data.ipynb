## 資料庫分(月)讀取

```python
def save_month_table(start_month,end_month_plus1,save_path,select_col,table,col):
    month_ls = pd.date_range(start=start_month, end=end_month_plus1, freq='M').strftime('%Y-%m')
    for ym in range(month_ls.shape[0]):
        sql = "select " + select_col + " from "+ table +" where " + col + " = '" + month_ls[ym] + "-01'"
        df = pd.read_sql(sql, conn)
        df.to_csv(save_path + '_' + month_ls[ym].replace('-', '') + '.csv')
        print(sql)
save_month_table('2016-07','2018-11','/project/edu-ofsn-wm/Data/CM_ACCOUNT_M','PROD_ID,GL_PROD_ID,DATA_YM,CUST_NO,MARKET_VAL_TWD,ACCT_BAL_TWD','nctu.CM_ACCOUNT_M','DATA_YM')
```

## 資料庫分區間(月)讀取

```python
def save_month_table(start_month,end_month_plus1,save_path,select_col,table,col):
    month_ls = pd.date_range(start=start_month, end=end_month_plus1, freq='M').strftime('%Y-%m')
    month_end = pd.date_range(start=start_month, end=end_month_plus1, freq='M').strftime('%Y-%m-%d')
    for ym in range(month_ls.shape[0]):
        sql = "select " + select_col + " from "+ table +" where " + col + " between '" + month_ls[ym] + "-01' and '" + month_end[ym] + "'"
        df = pd.read_sql(sql, conn)
        df.to_csv(save_path + '_' + month_ls[ym].replace('-', '') + '.csv')
#         print(sql)
save_month_table('2016-06','2018-11','/project/edu-ofsn-wm/Data/WITWO103_HIST1','WM_TXN_CODE,WM_PROD_CODE,UNIT,TXN_NAV_VAL,TXN_FX_RATE,TXN_DT,TXN_AMT,CUST_ID_CUST_NO','nctu.WITWO103_HIST','TXN_DT')
```

## 資料庫分(日)讀取

```python
def save_day_table(start_date,end_date_plus1,save_path,select_col,table,col):
    date = pd.date_range(start=start_date, end=end_date_plus1, freq='D').strftime('%Y-%m-%d')
    for ymd in range(date.shape[0]-1):
        sql = "select " + select_col + " from "+ table +" where " + col + " = '" + date[ymd] + "'"
        df = pd.read_sql(sql, conn)
        df.to_csv(save_path + '_' + date[ymd].replace('-', '') + '.csv')
#         print(sql)
save_day_table('2016-06','2016-07','/project/edu-ofsn-wm/Data/CM_ACCOUNT_M','PROD_ID,GL_PROD_ID,DATA_YM,CUST_NO,MARKET_VAL_TWD,ACCT_BAL_TWD','nctu.CM_ACCOUNT_M','DATA_YM')
```

## 統整月資料-->存成大檔

```python
def append_month_table(start_month,end_month_plus1,read_path):
    month_ls = pd.date_range(start=start_month, end=end_month_plus1, freq='M').strftime('%Y-%m')
    df = pd.read_csv(read_path + '_' + month_ls[0].replace('-', '') + '.csv')
    df.to_csv('/project/edu-ofsn-wm/Data/CM_CUSTOMER_M2.csv',mode='a',index=False)
    print(month_ls[0],df.shape)
    for ym in range(month_ls.shape[0]-1):
        df = pd.read_csv(read_path + '_' + month_ls[ym+1].replace('-', '') + '.csv')
        df.to_csv('/project/edu-ofsn-wm/Data/CM_CUSTOMER_M2.csv',mode='a',header=0,index=False)
        print(month_ls[ym+1],df.shape)
append_month_table('2016-06','2018-11','/project/edu-ofsn-wm/Data/CM_CUSTOMER_M2')
```

```python
### w103申購最近一次交易時間&前一年日期(with寫法)
sql = """
        with 
            cte1 as (select cust_id as cust_no, 
                            replace(wm_prod_code, ' ', '') as wm_prod_code, 
                            txn_dt,
                            (case when wms_txn_amt_twd is null then 1 else wms_txn_amt_twd end) as wms_txn_amt_twd,
                            deduct_cnt
                    from sinica.witwo103_hist  
                    where wm_txn_code='1' and txn_dt <= '{d_end}'),
            cte2 as (select distinct replace(wm_prod_code, ' ', '') as wm_prod_code
                    from sinica.witwo106
                    where replace(prod_detail_type_code, ' ','') in ('FNDF','FNDD')),
            cte3 as (select cte1.cust_no, cte1.wm_prod_code, cte1.txn_dt, cte1.wms_txn_amt_twd, cte1.deduct_cnt
                    from cte1 inner join cte2 on cte1.wm_prod_code=cte2.wm_prod_code order by cust_no limit 2742),
            cte4 as(select cust_no, txn_dt 
                    from cte3
                    where deduct_cnt in (0,1)),
            cte5 as (select cust_no, 
                     max(txn_dt) as max_txn_dt,
                     (max(txn_dt) - INTERVAL '1 year')::date as start_txn_dt
                     from cte4
                     group by cust_no)
            select cte3.cust_no, cte3.wm_prod_code, sum(cte3.wms_txn_amt_twd) as txn_amt
            from cte3 left join cte5 on cte3.cust_no=cte5.cust_no
            where cte3.txn_dt >= cte5.start_txn_dt and cte3.txn_dt <= cte5.max_txn_dt
            group by cte3.cust_no, cte3.wm_prod_code
        """.format(d_end='2019-06-30')
w103_recent_buy = pd.read_sql(sql, conn)
```
